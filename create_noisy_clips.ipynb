```python
def convert_mp3_to_wav(source_folder, target_folder):
    # Ensure target folder exists
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)

    # Convert each file
    for filename in os.listdir(source_folder):
        if filename.endswith(".mp3"):
            mp3_path = os.path.join(source_folder, filename)
            wav_path = os.path.join(target_folder, filename.replace(".mp3", ".wav"))

            # Load the mp3 file and export it as wav
            audio = AudioSegment.from_mp3(mp3_path)
            audio.export(wav_path, format="wav")
            print(f"Converted {filename} to WAV")
```

# Pitch Shift

```python
from audiotools import AudioSignal
```

```python
import torchaudio
```

```python
def pitch_shift(input_audio_path, pitch_shift_amount):
    

    if isinstance(pitch_shift_amount, torch.Tensor):
        pitch_shift_amount = pitch_shift_amount.long().item()

    sig = AudioSignal(input_audio_path)

    shifter = torchaudio.transforms.PitchShift(
        sig.sample_rate,
        n_steps=pitch_shift_amount, 
        bins_per_octave=12, 
        n_fft=512
    ) 
    sig.audio_data = shifter(sig.audio_data)
    return sig.audio_data, sig.sample_rate
```

```python
from audiotools import AudioSignal
import librosa
```

```python
def time_stretch(input_audio_path, time_stretch_amt):
    sig = AudioSignal(input_audio_path)
    sig.audio_data = torch.from_numpy(librosa.effects.time_stretch(
                    y=sig.audio_data.squeeze(0).numpy(),
                    rate=time_stretch_amt
                )).unsqueeze(0)
    return sig.audio_data, sig.sample_rate
```
